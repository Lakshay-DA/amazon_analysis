# Amazon E-commerce Sales Analysis project


### Import all libraries 


```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
sns.set_style("whitegrid")
```

## Import Data 


```python
file_path = "C:/Users/lxxsh/OneDrive/Documents/Amazon_project.csv"
df = pd.read_csv(file_path)
```

## Data Cleaning and Data Preprocessing


```python
df.head()
```

```python
df.info()
```

```python
df['rating_count'].isnull().sum()
```

```python
df.columns
```

```python
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
```

```python
df = df.copy()

df['rating_count'] = (
    df['rating_count']
    .astype(str)
    .str.replace(r'[^0-9]', '', regex=True)
)

df['rating_count'] = pd.to_numeric(df['rating_count'], errors='coerce')
df['rating_count'] = df['rating_count'].fillna(0)
```

```python
df['discounted_price'] = df['discounted_price'].astype(str)
df['discounted_price'] = df['discounted_price'].str.replace('â‚¹','')
df['discounted_price'] = df['discounted_price'].str.replace(',','')
df['discounted_price'] = pd.to_numeric(df['discounted_price'],errors='coerce')
```

```python
df['actual_price'] = df['actual_price'].astype(str)
df['actual_price'] = df['actual_price'].str.replace('â‚¹','')
df['actual_price'] = df['actual_price'].str.replace(',','')
df['actual_price'] = pd.to_numeric(df['actual_price'], errors='coerce')
```

```python
df['discount_percentage'] = df['discount_percentage'].astype(str)
df['discount_percentage'] = df['discount_percentage'].str.replace('%','')
df['discount_percentage'] = pd.to_numeric(df['discount_percentage'], errors='coerce')
```

# Analyst Insight

##### Missing values in the rating_count column were cleaned by extracting numeric values and replacing missing entries with zero to represent products with no customer reviews.

```python
df.isnull().sum()
```

```python
df = df.dropna(subset=['rating_count'])
df = df.dropna(subset=['rating'])
```

# Visualizations
## Rating distribution
###  Analysis Insight

```python
plt.figure(figsize=(8,5))
sns.histplot(df['rating'],bins=20)
plt.title('Distribution of Product Rating')
plt.xlabel('Rating')
plt.ylabel('Number of Products')
plt.show()
```

# Visualizations
## Discount vs Rating
#### Q1.Do higher discounts lead to better ratings?
### Analysis Insight
#### The scatter plot shows no strong relationship between discount percentage and product rating, suggesting that customer satisfaction is not significantly influenced by discount levels.

```python
plt.figure(figsize=(8,5))
sns.regplot(x='discount_percentage',y='rating',data=df,scatter_kws={'alpha':0.4})
plt.title('Discount Percentage vs Product Rating')
plt.xlabel("Discount percentage")
plt.ylabel("Rating")
plt.show()
```

# Visualizations
##   Discount Impact
#### Analysis Insight
#### Discounts do not significantly influence ratings.

```python
plt.figure(figsize=(10,6))

sns.regplot(
    x='discount_percentage',
    y='rating',
    data=df,
    scatter_kws={'alpha':0.3, 's':20},
    line_kws={'color':'red'}
)

plt.xticks(range(0,101,10))  # show every 10%

plt.grid(False)

plt.title("Discount vs Rating")
plt.xlabel("Discount Percentage")
plt.ylabel("Rating")

plt.show()
```
